# Task 1

```python
def sigmoid(x,the):
    s = np.dot(the,x.T)
    z = 1 / (1 + np.exp(-(s))) - 0.0000001
    return z
```

```python
def loss_func(x,y,the):
    m = len(x)
    y_ = sigmoid(x,the)
    return -(1/m)*np.sum(y*np.log(y_)+(1-y)*np.log(1-y_))
```

```python
def grad_desc(x,y,the,l_rate,Tepochs):
    m = len(x)
    k = [loss_func(x,y,the)]
    for i in range(Tepochs):
        y_ = sigmoid(x,the)
        for i in range(0,len(x.columns)):
            the[i] -= (l_rate/m) * np.sum((y_-y)*x.iloc[:,i])
        k.append(loss_func(x,y,the))
    return k , the
```

```python
def pred(x, y, the, l_rate, Tepochs):
    k, t = grad_desc(x,y,the,l_rate,Tepochs)
    y_ = sigmoid(x,the)
    for i in range(len(y_)):
        if y_[i] >= 0.5:
            y_[i] = 1
        else:
            y_[i] = 0
    y = list(y)
    accuracy = np.sum([y[i] == y_[i] for i in range(len(y))])/len(y)
    return k,accuracy
```
